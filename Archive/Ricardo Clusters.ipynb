### 3.1.3. KNNImputer

__`Step 57`__ Let's use the elbow method with K-means to try to identify the rigth number of clusters

```python
from sklearn.cluster import KMeans

ks = range(1, 11)
inertias = []

for k in ks:
    model = KMeans(n_clusters=k)
    model.fit(customer)
    inertias.append(model.inertia_)
```
