### 3.1.3. KNNImputer

__`Step 34`__ Apply KKNImputer to fill in the missing values in the variable "Year_Birth".

```python
from sklearn.impute import KNNImputer

df_numerical = df.drop(columns=['Churn', 'Name', 'Longevity', 'TypeTravel', 'RoomType'])

imputer = KNNImputer(n_neighbors=2, weights="uniform")
imputed_matrix = imputer.fit_transform(df_numerical)

df_cleaned = pd.DataFrame(data=imputed_matrix, columns=df_numerical.columns, index=df_numerical.index)
```

__`Step 35`__ Confirm we have no missing values after the imputation.

```python
df_cleaned['Year_Birth'].describe()
```

```python
df_cleaned.shape
```

__`Step 37`__ Join the new imputed values with our original dataframe.

```python
df.columns
```

```python
df_numerical = df[['Churn', 'Name', 'Longevity', 'TypeTravel', 'RoomType']]
```

__`Step 56`__ Select the variables for the Customer Characteristics segmentation. We are going to use: 'Age', 'TypeTravel' and 'RoomType'. For that purpose, we need to create a specific dataframe with those variables only. Because they have different meausures, we will remove them from the scaled_df dataframe.

```python
customer = scaled_df[['Age', 'TypeTravel_leisure', 'RoomType_single', 'RoomType_suite']]
customer
```

<a id='customer_clusters'></a>

### 4.1.1. Identify the rigth number of clusters

__`Step 57`__ Let's use the elbow method with K-means to try to identify the rigth number of clusters

```python
from sklearn.cluster import KMeans

ks = range(1, 11)
inertias = []

for k in ks:
    model = KMeans(n_clusters=k)
    model.fit(customer)
    inertias.append(model.inertia_)
```

__`Step 58`__ Now let's plot the inertias into a figure

```python
plt.plot(ks, inertias)
plt.xlabel('number of clusters')
plt.ylabel('inertia')
plt.xticks(ks)
plt.show()
```

We can see from the graph that 3 seems to be a good number of clusters, since this is the point where the inertia (or the sum squared of errors) begins to decrease more slowly. However, it is always a good practice to see if the nearest values can be better solutions.

__`Step 59`__ Before applying K-means into our data to test different K values, let's use another method to try to identify the right number of cluster, using hierarchical clustering method, and plot a dendogram

```python
from sklearn.datasets import load_iris
from sklearn.cluster import AgglomerativeClustering
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram , linkage

Z = linkage(customer, method = 'ward')
 
dendro = dendrogram(Z)
plt.title('Dendrogram')
plt.ylabel('Euclidean distance')
plt.show()
```

As per the dendogram the ideal number of cluster seems to be 3, as well. However, as mentioned previously, we will try different K's to see the result
