**Quick transformations**

* Longevity: yes = 1, no = 0 <br>
* Churn: churn = 1, nochurn = 0 <br>
* TypeTravel: business = 1, leisure = 0 <br>
* RoomType: DoubleRoom: 0=single, 1=double; Suite: 1=yes, 0=no

```python
df['Longevity'] = df['Longevity'].replace({'y':1,'yes':1,'no':0}).astype(int)
df['Churn'] = df['Churn'].replace({'churn':1,'nochurn':0}).astype(int)
df['TypeTravel'] = df['TypeTravel'].replace({'business':1,'leisure':0}).astype(int)
df['Suite'] = df['RoomType'].replace({'suite':1,'double':0,'single':0}).astype(int)
df['DoubleRoom'] = df['RoomType'].replace({'suite':0,'double':1,'single':0}).astype(int)
df['Year_Birth'] = df['Year_Birth'].astype(int)
df['Wifi'] = df['Wifi'].replace(6,5)
df = df.drop(['RoomType'],axis=1)
df['Age'] = date.today().year - df['Year_Birth']
```
